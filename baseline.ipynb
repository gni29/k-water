```python
import seaborn as sns
# 서브플롯 설정: 한 줄에 2개씩 배치
num_cols = 2  # 한 줄에 표시할 그래프 개수
num_vars = len(a_tr.columns)  # 전체 변수 개수
num_rows = (num_vars + num_cols - 1) // num_cols  # 필요한 행 수 계산

fig, axes = plt.subplots(num_rows, num_cols,figsize=(20, num_rows * 4))  # 전체 그래프 크기
axes = axes.flatten()  # 2D 배열을 1D로 펼치기

# 변수별로 그래프 그리기
for i, column in enumerate(a_tr.columns):

    sns.lineplot(data=a_tr[column], ax=axes[i])
    axes[i].set_title(f'{column} Line Plot')
    axes[i].set_xlabel('Index')
    axes[i].set_ylabel(column)

# 빈 서브플롯 제거
for i in range(num_vars, len(axes)):
    fig.delaxes(axes[i])
plt.subplots_adjust(hspace=0.5, wspace=0.3)
plt.show()
```
