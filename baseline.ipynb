```python
import seaborn as sns
# ì„œë¸Œí”Œë¡¯ ì„¤ì •: í•œ ì¤„ì— 2ê°œì”© ë°°ì¹˜
num_cols = 2  # í•œ ì¤„ì— í‘œì‹œí•  ê·¸ëž˜í”„ ê°œìˆ˜
num_vars = len(a_tr.columns)  # ì „ì²´ ë³€ìˆ˜ ê°œìˆ˜
num_rows = (num_vars + num_cols - 1) // num_cols  # í•„ìš”í•œ í–‰ ìˆ˜ ê³„ì‚°

fig, axes = plt.subplots(num_rows, num_cols,figsize=(20, num_rows * 4))  # ì „ì²´ ê·¸ëž˜í”„ í¬ê¸°
axes = axes.flatten()  # 2D ë°°ì—´ì„ 1Dë¡œ íŽ¼ì¹˜ê¸°

# ë³€ìˆ˜ë³„ë¡œ ê·¸ëž˜í”„ ê·¸ë¦¬ê¸°
for i, column in enumerate(a_tr.columns):

    sns.lineplot(data=a_tr[column], ax=axes[i])
    axes[i].set_title(f'{column} Line Plot')
    axes[i].set_xlabel('Index')
    axes[i].set_ylabel(column)

# ë¹ˆ ì„œë¸Œí”Œë¡¯ ì œê±°
for i in range(num_vars, len(axes)):
    fig.delaxes(axes[i])
plt.subplots_adjust(hspace=0.5, wspace=0.3)
plt.show()
```
